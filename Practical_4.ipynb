```python
import geemap.geemap as map
```

```python
import ee
```

```python
ee.Authenticate()
```

```python
ee.Initialize(project = 'ee-nalinapk17')
```

```python
m=map.Map()
m
```

```python
India = ee.FeatureCollection('projects/ee-nalinapk17/assets/India_states')
India
```

```python
m.addLayer(India,{},'India') #dislpaying the shapefile in the interactive map
m
```

```python
India_Districts = ee.FeatureCollection('projects/ee-nalinapk17/assets/India_Districts')
m.addLayer(India_Districts,{},'India_Districts') #dislpaying the shapefile in the interactive map
m
```

```python
IndiaDEM = ee.Image("USGS/SRTMGL1_003").clip(India_Districts)
Visualize = {'min': 0, 'maz': 8700, "palette": ["#003300", "#006600",  "#009933",  "#33CC33", "#66FF66", "#E5FFCC",  "#CC9900",  "#996633",  "#804000",  "#662A00",  "#B2B2B2", "#D8D8D8", "#F0F0F0",  "#F5F5F5",  "#FFFFFF" ]}
m.addLayer(IndiaDEM, Visualize, 'District_dem')
m
```

```python
treeCover = ee.ImageCollection("MODIS/006/MOD44B").filterDate("2019-01-01","2019-12-31").select("Percent_Tree_Cover").mean().clip(India_Districts)
Visualize_Tree = {"min":0, "max":1, 'palette': ['bbe029', '0a9501', '074b03']}
m.addLayer(treeCover, Visualize_Tree, 'treeCover')
m
```

```python
import os
```

```python
out_dir = os.path.join(os.path.expanduser('~'), 'Downloads')
India_treecover = os.path.join(out_dir, 'India_treecover.csv')
if not os.path.exists(out_dir):
    os.makedirs(out_dir)
map.zonal_stats(treeCover, India_Districts, India_treecover, statistics_type='MEAN', scale=250)
```
